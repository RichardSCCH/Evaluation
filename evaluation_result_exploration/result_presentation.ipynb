```python
ncs = [pd.read_csv('./ncs/code.original_subtoken', header=None)]
ncs.append(pd.read_json("./ncs/code2jdoc_beam.json").T)
ncs.append(pd.read_csv('./ncs/javadoc.original', header=None))

ncs = pd.concat(ncs, axis=1, join='inner')
ncs.columns = ["method", "predicted_summary", "reference_summary"]
ncs['method'] = ncs['method'].astype(str)
ncs['predicted_summary'] = ncs['predicted_summary'].astype(str)
ncs['reference_summary'] = ncs['reference_summary'].astype(str)
ncs_sample = ncs.sample(3)
ncs_sample
```

```python
leclair = pd.read_csv("./code-gnn-gru/predict-codegnngru.tsv", sep="\t", header=None)
comments = pd.read_csv("./code-gnn-gru/comments.tsv", sep="\t", header=None)
comments.columns = ["id", "reference_summary"]
leclair.columns = ["id", "method", "predicted_summary"]
leclair = pd.merge(leclair, comments, how="left", on="id")

leclair_sample = leclair.sample(3)
leclair_sample
```

```python
leclair.describe(exclude=np.number)
```

```python
def create_pred_ref_plot(data: pd.DataFrame, title, axVal):

    pred_len = data["predicted_summary"].str.split().apply(len)
    pred_len = pd.concat([pred_len, data["reference_summary"].str.split().apply(len)], axis=1)

    df = pred_len.melt(var_name='summary_type', value_name='length')

    summary_len_plot = sns.countplot(x='length', hue='summary_type', data=df, ax=axVal)
    summary_len_plot.set_xlabel("Length of summaries")
    summary_len_plot.set_ylabel("Length of candidate/reference summaries")
    summary_len_plot.set(title=title)
```

```python
sns.set(rc = {'figure.figsize':(20,8)})
fig, axs = plt.subplots(ncols=3, sharey=True)

leclair["predicted_summary"] = leclair["predicted_summary"].str.replace(r'<((s)|(/s)|(NULL))>', '')

create_pred_ref_plot(leclair, 'code+gnn+gru', axs[0])
create_pred_ref_plot(ncs, 'NCS', axs[1])
create_pred_ref_plot(rencos, 'Rencos', axs[2])
```

# Print metrics of the results

```python
data = pd.read_csv('./evaluation-result.csv', index_col=0)
data[['codegnngru_bleu', 'ncs_bleu', 'rencos_bleu', 'codegnngru_rouge', 'ncs_rouge', 'rencos_rouge']].describe()
```

```python
sns.set(rc = {'figure.figsize':(20,15)})
fig, axs = plt.subplots(nrows=2, sharey=True)

sns.set(font_scale = 4)

bleu = sns.violinplot(x="variable", y="value", data=pd.melt(data[['codegnngru_bleu', 'ncs_bleu', 'rencos_bleu']]), ax=axs[0])
bleu.set(xlabel='Model', ylabel='BLEU score')

rouge = sns.violinplot(x="variable", y="value", data=pd.melt(data[['codegnngru_rouge', 'ncs_rouge', 'rencos_rouge']]), ax=axs[1])
rouge.set(xlabel='Model', ylabel='ROUGE-L score')
```
